# Data Visualization


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("netflix_titles.csv")
df.columns = df.columns.str.strip()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['release_year'], bins=20)
plt.title("Histogram of Release Years")
plt.show()
```

This histogram shows the distribution of Netflix titles based on their release year.
Most content is concentrated after the year 2000, with a sharp rise after 2015.
Very few titles exist before 1980, indicating older content is limited.
Meaningful Insight

```python
plt.figure(figsize=(10,5))
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries with Most Content")
plt.show()
```

This bar chart displays the top 10 countries contributing the highest number of titles on Netflix.
The United States dominates the content library, followed by India and the UK.
The USA has significantly more titles compared to other countries, showing a large gap.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='type', y='release_year', data=df)
plt.title("Box Plot: Release Year vs Type")
plt.show()
```

This box plot compares the release year distribution for Movies and TV Shows.
TV shows tend to have more recent release years compared to movies.
Some very old movies appear as outliers, indicating Netflix includes a few classic films.

```python
df['content_age'] = 2025 - df['release_year']
plt.figure(figsize=(8,5))
sns.scatterplot(x='release_year', y='content_age', data=df)
plt.title("Scatter Plot: Release Year vs Content Age")
plt.show()
```

This scatter plot shows the relationship between release year and content age.
Newer titles have lower content age, while older titles have higher content age.
A few very old titles appear far from the main cluster, representing rare classic content.

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['added_year'] = df['date_added'].dt.year
plt.figure(figsize=(8,5))
df['added_year'].value_counts().sort_index().plot(kind='line')
plt.title("Line Plot: Titles Added Each Year")
plt.show()
```

This line plot shows the number of titles added to Netflix each year.
There is a rapid increase in content additions after 2015, peaking around 2019–2020.
Some years show sudden drops, possibly due to incomplete data or reduced content releases.

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Heatmap of Missing Values")
plt.show()
```

This heatmap highlights missing values across different columns in the dataset.
Columns like director, cast, and country contain many missing values.
Some records have multiple missing fields, indicating incomplete metadata.

```python
sns.pairplot(df[['release_year', 'content_age']].dropna())
plt.show()
```


This pair plot visualizes relationships between numerical variables like release year and content age.
Release year and content age show a strong inverse relationship (newer content has lower age).
Few points lie outside the main pattern, representing very old content.

```python
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=df)
plt.title("Count Plot: Movies vs TV Shows")
plt.show()
```

This count plot compares the number of Movies and TV Shows available on Netflix.
Movies are significantly more than TV Shows in the dataset.
The imbalance indicates Netflix’s catalog is movie-heavy

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='type', y='release_year', data=df)
plt.title("Violin Plot: Release Year by Type")
plt.show()
```

This violin plot shows the density distribution of release years for Movies and TV Shows.
TV shows have a tighter distribution around recent years, while movies are spread across a wider range.
Movies show longer tails, meaning Netflix includes some older films

 Bivariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.countplot(x="rating", hue="type", data=df)
plt.title("Type vs Rating Distribution")
plt.xticks(rotation=45)
plt.show()
```

This count plot shows the distribution of Netflix content ratings, separated by Movies and TV Shows.
Movies dominate most rating categories, while TV Shows appear more in TV-MA and TV-14 categories.
Some ratings have very few titles, indicating rare content categories

```python
top_countries = df['country'].value_counts().head(10).index

plt.figure(figsize=(10,5))
sns.countplot(y="country", hue="type",
              data=df[df['country'].isin(top_countries)])
plt.title("Top 10 Countries: Movies vs TV Shows")
plt.show()
```

This plot compares Movies and TV Shows across the top 10 countries producing Netflix content.
The USA has the highest number of both Movies and TV Shows, followed by India and the UK.
Some countries contribute mostly movies and very few TV shows.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="type", y="release_year", data=df)
plt.title("Release Year Distribution by Type")
plt.show()
```

This box plot shows how the release years differ between Movies and TV Shows.
TV shows tend to have more recent release years compared to movies.
Some movies appear as outliers with very old release years

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['added_year'] = df['date_added'].dt.year

plt.figure(figsize=(10,5))
sns.countplot(x="added_year", hue="type", data=df)
plt.title("Titles Added Each Year by Type")
plt.xticks(rotation=45)
plt.show()
```

This plot shows the number of Movies and TV Shows added to Netflix each year.
Netflix rapidly increased content additions after 2016, especially movies.
Some years show fewer additions, possibly due to incomplete records.

```python
df['duration_num'] = df['duration'].str.extract('(\d+)').astype(float)

plt.figure(figsize=(8,5))
sns.boxplot(x="type", y="duration_num", data=df)
plt.title("Duration Distribution by Type")
plt.show()
```

This plot compares the duration distribution between Movies and TV Shows.
Movies generally have higher duration values, while TV shows have smaller values because duration represents seasons.
Some movies have unusually long durations, appearing as outliers.

## *Multivariate*

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Heatmap of Missing Values in Netflix Dataset")
plt.show()
```

This heatmap shows missing data patterns across all columns in the Netflix dataset.
Columns like director, cast, and country contain more missing values compared to others.
Some rows have multiple missing fields, indicating incomplete metadata entries.

```python
df['content_age'] = 2025 - df['release_year']
sns.pairplot(df[['release_year', 'content_age']].dropna())
plt.show()
```

This pair plot visualizes relationships between numerical features like release year and content age.
A strong inverse relationship is visible: newer titles have lower content age.
A few points represent very old releases, far from the main cluster.

```python
top_countries = df['country'].value_counts().head(5).index
plt.figure(figsize=(12,6))
sns.countplot(
    data=df[df['country'].isin(top_countries)],
    x="country",
    hue="rating"
)
plt.title("Ratings Distribution Across Top Countries")
plt.xticks(rotation=45)
plt.show()
```

This plot shows how content ratings vary across the top Netflix-producing countries.
Different countries have different dominant rating categories (e.g., US has more TV-MA content).
Some countries show very few titles in certain rating groups.

CONCLUSION
### Based on the exploratory data analysis of the Netflix Movies and TV Shows dataset, it is evident that Netflix’s content library is largely dominated by movies, while TV shows have been increasing steadily over recent years. The platform experienced rapid growth in content additions after 2015, highlighting Netflix’s strong focus on expanding its streaming catalog worldwide. The analysis also shows that the United States contributes the highest number of titles, followed by countries like India and the United Kingdom, indicating that Netflix content is concentrated around a few major production regions.

## Furthermore, genre and rating-based insights reveal that Drama and International content are among the most common categories, reflecting Netflix’s emphasis on widely appealing storytelling. The rating distribution suggests that a significant portion of Netflix content targets mature and teen audiences, especially through categories like TV-MA and TV-14. Overall, the EDA concludes that Netflix prioritizes modern, diverse, and globally expanding entertainment, with newer content being added more frequently while older titles appear only occasionally as outliers.
